```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
signals = ['lq_lq', 'lq_ta', 'ta_ta', 'zp_tau_tau']
```

```python
g = 2

case = "woRHC"

fig = plt.figure(figsize = (5,6))

for signal in signals:
    data = pd.read_csv(f'xs_signals_{case}/{signal}/Cross_Sections/XS_Matriz.csv', index_col = 0)
    data.index = [float(i) for i in data.index]
    data.columns = [float(i) for i in data.columns]
    
    X = data.columns
    Y = [data[m][g]for m in data.columns]
    etiqueta = signal.replace('lq_ta','sLQ').replace('lq_lq','dLQ').replace('ta_ta','non-res').replace('zp_tau_tau', r'$Z^{\prime}$')
    
    plt.plot(X/1000,Y, label = etiqueta)
    
plt.xlim(1,2.5)
plt.xlabel(r'$M_{U}$ [TeV]', loc= 'right', fontsize = 12)
plt.ylabel(r'$\sigma $ (pb)', loc = 'top', fontsize = 12)
plt.yscale('log')
plt.grid()
plt.legend()
plt.title(r'$\mathbf{\sqrt{s} = 13 TeV, g_{U} =' + f'{g}' + r', \beta_{R} = 0}$', fontweight='bold', loc = 'right')
plt.show()
```
