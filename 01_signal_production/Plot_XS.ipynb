```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
g = 2
case = "woRHC"

fig = plt.figure(figsize = (5,5))

# Dictionary of labels
signals = {
    'lq_lq': "dLQ", 
    'lq_ta': 'sLQ', 
    'ta_ta': 'non-res', 
    'zp_tau_tau': r'$Z^{\prime} \rightarrow \tau \tau$'
    }

for signal in signals.keys():
    # Read xs data
    path_tp_xs = f'xs_signals_{case}/{signal}/XS_Matrix.csv'
    data = pd.read_csv(path_tp_xs, index_col = 0)

    # cast index and columns to float
    data.index = [float(i) for i in data.index]
    data.columns = [float(i) for i in data.columns]
    
    # Select data from a gived g
    X = data.columns
    Y = [data[m][g]for m in data.columns]
    
    # Plot
    plt.plot(X/1000,Y, label = signals[signal])
    
plt.xlim(1,2.5)
plt.xlabel(r'$M_{U}$ [TeV]', loc= 'right', fontsize = 12)
plt.ylabel(r'$\sigma $ (pb)', loc = 'top', fontsize = 12)
plt.yscale('log')
plt.grid()
plt.legend()
plt.title(r'$\mathbf{\sqrt{s} = 13 TeV, g_{U} =' + f'{g}' + r', \beta_{R} = 0}$', fontweight='bold', loc = 'right')
plt.show()
```
