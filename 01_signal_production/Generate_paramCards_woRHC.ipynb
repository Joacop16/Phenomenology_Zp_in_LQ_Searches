First, consider the case with coupling only to left-handed third-generation fermions.

Assume that the leptoquark interaction with third gen fermions is given by the following Lagrangian:
$$
    \mathcal{L}_U =  \frac{g_U}{\sqrt{2}}  U^\mu \left[ \beta_L^{b\tau} \bar q^{3}_L \gamma_u \ell^3_L  \right] + \textrm{h.c.}
$$
while the interaction for $Z'$ is given by:
$$
    \mathcal{L}_{Z'} =  \frac{g_{Z'}}{2\sqrt 6}  Z'^\mu \left[  \zeta^{b\,b}_L \bar q_L^3 \gamma _u q_L^3 -3 \zeta_\ell^{\tau\tau} \bar\ell ^3_L\gamma_\mu \ell^3_L \right] 
$$
with $\zeta^{b\,b}_L =  \zeta_\ell^{\tau\tau} = \beta_L^{b\tau} = 1$.
In the case of 4321 model $M_{Z'}\approx \frac 1{\sqrt 2}M_U$, and $g_{Z'}\approx g_U$.

```python
import os
import numpy as np
import subprocess
import shutil

MG5_Path = os.path.join(os.sep, 'Collider', 'MG5_aMC_v3_1_0', 'bin', 'mg5_aMC')
MG5_Path = os.path.relpath(MG5_Path, os.getcwd())

mass_step = 500 # GeV
g_u_step = 1/4 # 
M_U = np.arange(1000, 3500 + mass_step, mass_step)
g_U = np.arange(0.5, 3.5 + g_u_step, g_u_step)

M_zp = M_U/np.sqrt(2)
g_Zp = g_U

UFO_name = "Mod2_VLQ_UFO"
case= "woRHC"

output_folders = {}

header = f"""
import model {UFO_name} 
generate vlq > all all
add process zp > all all
"""
file = open(f'Paramcards_{case}.mg5','w')
file.write(header)

# Ensure that tmp folder is empty
shutil.rmtree('tmp', ignore_errors=True)
os.makedirs('tmp', exist_ok=True)
for mass in M_U:
    output_folders[mass] = {}
    for g in g_U:
        output_folder = "tmp/M{}_gU{:.4f}".format(mass, g).replace('.', '_')
        file.write(f"output {output_folder} -nojpeg\n")
        output_folders[mass][g] = output_folder
file.close()
```

```python
try :
    subprocess.run([MG5_Path, f'Paramcards_{case}.mg5'], stdout=subprocess.DEVNULL, check=True)
except :
    print("Madgraph is not installed in {}".format(MG5_Path))
    
```

Remember the full lagrangian in the feynrules model is for the leptoquark and the $Z'$ boson is given by:
$$
\begin{aligned}
\mathcal{L}_{U_1}= & -\frac{1}{2} U_{1 \mu \nu}^{\dagger} U_1^{\mu \nu}+M_U^2 U_{1 \mu}^{\dagger} U_1^\mu \\
& -i g_s\left(1-\kappa_U\right) U_{1 \mu}^{\dagger} T^a U_{1 \nu} G^{a \mu \nu} \\
& -i g_Y \frac{2}{3}\left(1-\tilde{\kappa}_U\right) U_{1 \mu}^{\dagger} U_{1 \nu} B^{\mu \nu} \\
& +\frac{g_U}{\sqrt{2}}\left[U_1^\mu\left(\beta_L^{i j} \bar{q}_L^i \gamma_\mu \ell_L^j+\beta_R^{i j} \bar{d}_R^i \gamma_\mu e_R^j\right)+\text { h.c. }\right],
\end{aligned}
$$
and 
$$
\begin{aligned}
\mathcal{L}_{Z^{\prime}}= & -\frac{1}{4} Z_{\mu \nu}^{\prime} Z^{\prime \mu \nu} \\
& +\frac{1}{2} M_{Z^{\prime}}^2 Z_\mu^{\prime} Z^{\prime \mu} \\
& +\frac{g_{Z^{\prime}}}{2 \sqrt{6}} Z^{\prime \mu}\left(\zeta_q^{i j} \bar{q}_L^i \gamma_\mu q_L^j+\zeta_u^{i j} \bar{u}_R^i \gamma_\mu u_R^j\right. \\
& \left.+\zeta_d^{i j} \bar{d}_R^i \gamma_\mu d_R^j-3 \zeta_{\ell}^{i j} \bar{\ell}_L^i \gamma_\mu \ell_L^j-3 \zeta_e^{i j} \bar{e}_R^i \gamma_\mu e_R^j\right),
\end{aligned}
$$
 With third gen preferential coupling 
 $$
\begin{array}{rlr}
\beta_L & =\left(\begin{array}{ccc}
0 & 0 & \beta_L^{13} \\
0 & 0 & \beta_L^{23} \\
0 & \beta_L^{32} & \beta_L^{33}
\end{array}\right), \quad & \beta_R=\operatorname{diag}\left(0,0, \beta_R^{33}\right), \\
\zeta_{\ell} & =\left(\begin{array}{ccc}
0 & 0 & 0 \\
0 & \zeta_{\ell}^{22} & \zeta_{\ell}^{23} \\
0 & \left(\zeta_{\ell}^{23}\right)^* & \zeta_{\ell}^{33}
\end{array}\right), \quad & \zeta_e=\operatorname{diag}\left(0, \zeta_e^{22}, \zeta_e^{33}\right), \\
\zeta_q & =\operatorname{diag}\left(\zeta_q^{l l}, \zeta_q^{l l}, \zeta_q^{33}\right). &
\end{array}
$$
where $\zeta^{b\,b}_q =  \zeta_\ell^{\tau\tau} = \beta_L^{b\tau} = 1$ and all other couplings are zero.

```python
nevts = 1000 # number of events

# VLQ parameters
betal33 = 1
betar33 = 0

# Z' parameters
zetaq33 = 1
zetal33 = 1
zetaru33 = 0
zetard33 = 0
zetare33 = 0

# All other are already setted to 0 in the model

kappazp = 0 # coupling between Z' and VLQ

for i, mass in enumerate(M_U):
    for j, g in enumerate(g_U):
        with open(os.path.join(output_folders[mass][g], "CalculateDecayWidths.mg5"), 'w') as file:
            output_folder_abs = os.path.abspath(output_folders[mass][g])
            file.write(f"launch {output_folder_abs} -i\n")
            file.write("calculate_decay_widths\n")
            file.write(f"set nevents {nevts}\n")
            # setup leptoquark parameters
            file.write(f"set MVLQ {mass}\n")
            file.write(f"set GU {g}\n")
            file.write(f"set BETAL33 {betal33}\n")
            file.write(f"set BETAR33 {betar33}\n")
            # setup Z' parameters
            file.write(f"set MZP {M_zp[i]}\n")
            file.write(f"set GZP {g_Zp[j]}\n")
            file.write(f"set ZETAQ33 {zetaq33}\n")
            file.write(f"set ZETAL33 {zetal33}\n")
            file.write(f"set ZETARU33 {zetaru33}\n")
            file.write(f"set ZETARD33 {zetard33}\n")
            file.write(f"set ZETARE33 {zetare33}\n")
            # setup coupling between Z' and VLQ
            file.write(f"set KAPPAZP {kappazp}\n")
            file.write("exit")
        try :
            # Run Madgraph to calculate decay widths
            subprocess.run([MG5_Path, os.path.join(output_folder_abs, "CalculateDecayWidths.mg5")], check=True)
            # Create folder Paramcards_case if it does not exist
            os.makedirs(f"Paramcards_{case}", exist_ok=True)
            # Save new param_card.dat
            shutil.copyfile(
                os.path.join(output_folder_abs, "Events", "run_01", "param_card.dat"),
                os.path.join("Paramcards_{}/param_card_M{}_gU{:.4f}.dat".format(case, mass, g).replace('.', '_'))
            )
        except :
            print("Madgraph is not installed in {}".format(MG5_Path))

# clean tmp folder
shutil.rmtree('tmp', ignore_errors=True)
```
