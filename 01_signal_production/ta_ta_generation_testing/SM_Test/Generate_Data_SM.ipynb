```python
generation_options = {'e_e_virtual_contribution':  'generate p p > e+ e- $ z', 
                      'e_e_real_contribution': 'generate p p > z, z > e+ e-',
                      'e_e_total_contribution': 'generate p p > e+ e-'}
```

```python
seeds = []
def semilla():
    seed = np.random.randint(1,5000)
    if (seed in seeds):
        return semilla()
    else: seeds.append(seed)
    return seed

def generate_data(process):
    
    script_lines = [f'import model sm']    
    script_lines.append(generation_options[process])
    script_lines.append(f'output {os.path.join(os.getcwd(), process)} -nojpeg')
    script_lines.append(f'launch {os.path.join(os.getcwd(), process)} -m')
    script_lines.append('16')
    script_lines.append('done')
    script_lines.append(f'set iseed {semilla()}')
    script_lines += ['set nevents 100000', 'set cut_decays True', 'set ptb 30', 'set ptj 20', 'set ptl 20', 'set etab 2.5', 'set pt_min_pdg {15: 30}', 'set eta_max_pdg {15: 2.5}', 'set mmll 20']

    script = open (f'script_{process}.txt','w')
    for line in script_lines: script.write(line + '\n')
    script.close()
    !{MG5_Path} script_{process}.txt
```

```python
with Pool(3) as p: p.map(generate_data, generation_options.keys())
```
